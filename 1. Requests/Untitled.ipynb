```python
urll= 'https://jsonplaceholder.typicode.com/posts'
```

```python
def raise_for_statu(response):
    if response.status_code == 200:
        print('Success!!')
    else:
        print('Unsuccessfull!!')

    
    
```

```python
responses = requests.get(url)
raise_for_statu(responses)
```

```python
try:
    response= requests.get(urll)
    response.raise_for_status()
except Exception as e:
    print(e)
else:
    response_code = response.status_code
    print(response_code)

    if response_code == 200:
        print('successfull request!!')
        content=response.json()

        for i in range(4):
            print(f'\nresponse {i+1}')
            print(dict(content[i]))
    else:
        print('uncessful!!')

            
```

```python
data= {
    'user':'rohanp',
    'pass':'1234',
    'login':'yes'
}
try:
    response= requests.post(urll,data=data)
    response.raise_for_status()
except Exception as e:
    print(e)
else:
    response_code = response.status_code
    print(response_code)

    if response_code == 201:
        print('successfull request!!')
        content=response.json()
        data=response.json()
        print(data)
    else:
        print('uncessful!!')

            
```
